```python
def categorize_age(age):
    if 18 <= age <= 30:
        return '18-30'
    elif 31 <= age <= 50:
        return '31-50'
    elif age >= 51:
        return '51+'
    else:
        return 'Unknown'

# Apply the function to create a new column
marketing_df['Age_Group'] = marketing_df['Age'].apply(categorize_age)
```
